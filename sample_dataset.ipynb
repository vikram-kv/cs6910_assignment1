```python
from keras.datasets import fashion_mnist
import numpy, pandas
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

class_names_dict = {0 : 'T-shirt/top',  1	: 'Trouser', 2	: 'Pullover', 3	: 'Dress', 
4	: 'Coat', 5	: 'Sandal', 6	: 'Shirt', 7	: 'Sneaker', 8	: 'Bag', 9	: 'Ankle boot'}
```

```python
class_labels = set()

for c in y_train:
    class_labels.add(c)

print(class_labels)
num_classes = len(class_labels)
```

```python
# Representative images for each class
class_to_image = [[] for i in range(num_classes)]
for idx in range(x_train.shape[0]):
    class_to_image[y_train[idx]] += [idx]

figure, axis = plt.subplots(2, 5, figsize=(20,8))

for c in class_labels:
    i, j = c//5, c%5
    axis[i][j].set_title(f'Class = {c} ({class_names_dict[c]})')
    axis[i][j].imshow(x_train[class_to_image[c][0]], cmap='Greys')
    axis[i][j].axis('off')

plt.show()
```
